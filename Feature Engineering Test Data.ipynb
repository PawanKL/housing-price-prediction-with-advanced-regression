## Numerical Features

````python
## Numerical variables containing missing values
numerical_with_nan=[feature for feature in data.columns if data[feature].isnull().sum()>0 and data[feature].dtypes!='O']

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(data[feature].isnull().mean() * 100,4)))
````

### Handling Missing Values
``` Replacing LotFrontage null value with median value of Lotfrontage ```

``` Replacing all basement features with 0 beacause in this case missing values could be having no basement ```

``` Replacing all Garage features with 0 beacause in this case missing values means there is no garage which is true with           GarageCars have zero value ```

``` Replacing MasVnrArea with 0 ```

````python
LotFrontage_median = data['LotFrontage'].median()
data["LotFrontage"] = data['LotFrontage'].fillna(LotFrontage_median)
````

````python
for feature in ['GarageYrBlt', 'GarageArea', 'GarageCars']:
    data[feature] = data[feature].fillna(0)
````

````python
for feature in ['BsmtFinSF1', 'BsmtFinSF2', 'TotalBsmtSF', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath']:
    data[feature] = data[feature].fillna(0)
````

````python
data['MasVnrArea']   = data['MasVnrArea'].fillna(0)
````

````python
data.head(10)
````

## Log Transformation
### with EDA these features were skewed, so applying log transformation

````python
skewed_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea']

for feature in skewed_features:
    data[feature]=np.log(data[feature])
````

````python
data[skewed_features].head(10)
````

## Categorical Features

### Handling Special Cases
``` MSZoning contains 4 missing values in Test Set, but no missing value in Training Set ```

``` Utilities contains 2 missing values in Test Set, but no missing value in Training Set ```

``` Functional contains 2 missing values, but no missing value in Training Set ```

````python
## Handling MSZoning By replacing null value with mode
print(data['MSZoning'].unique())
MSZoning_mode_value = data['MSZoning'].mode()[0]
data['MSZoning'] = data['MSZoning'].fillna(MSZoning_mode_value)

## Droping Utilities feature as all records are AllPub with one NoSeWa (in training set), 2 NA in Test Set
data = data.drop(['Utilities'], axis=1)

## Handling Functional by replacing null value with mode
print(data['Functional'].unique())
Functional_mode_value = data['Functional'].mode()[0]
data['Functional'] = data['Functional'].fillna(Functional_mode_value)
````

````python
categorical_with_nan=[feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtypes=='O']

for feature in categorical_with_nan:
    print("{}: {}% missing values".format(feature,np.round(data[feature].isnull().mean()*100,4)))
````

````python
data_copy = data.copy()
data_copy[categorical_with_nan]=data_copy[categorical_with_nan].fillna('Missing')
data = data_copy
````

````python
data[categorical_with_nan].isnull().sum()
````

````python
categorical_features = [feature for feature in data.columns if data[feature].dtype == 'O']
print(len(categorical_features))
print(categorical_features)
````

### Handling Rare Categorical Variables
```if a cateogry occurs less then one percent then it can be set Rare_Cat```

````python
for feature in categorical_features:
    print(data[feature].unique())
    temp=data.groupby(feature)['Id'].count()/len(data)
    print(temp)
    temp_df=temp[temp>0.01].index
    data[feature]=np.where(data[feature].isin(temp_df),data[feature],'Rare_Cat')
````

````python
for feature in categorical_features:
    print(feature)
    data[feature] = LabelEncoder().fit_transform(data[feature])
````
