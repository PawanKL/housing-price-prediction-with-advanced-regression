```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error
import pickle
import joblib
```

```python
pd.pandas.set_option('display.max_columns', None)
np.set_printoptions(suppress=True)
model_path = ".\\saved models\\"
filename = './Test/X_test.csv'
```

```python
data = pd.read_csv(filename)
```

```python
data.head()
```

```python
print(type(data))
print(data.shape)
```

```python
X = data.drop('Id', axis=1)
print(type(X))
print(X.shape)
```

```python
model_file = model_path + 'Elastic_X_train.pkl'
with open(model_file, 'rb') as file:
    model = joblib.load(file)
print(model)
predictions = model.predict(X)
print(predictions)
```

```python
real_val = np.exp(predictions)
print(real_val)
```

```python
print(np.log(real_val[0]))
```

```python
submit_data = pd.DataFrame({'Id': data['Id'], 'SalePrice': real_val})
submit_data.head()
```

```python
submit_data.to_csv('submission_X_Elastic.csv', index=False)
```
